```python
%%writefile code.js

const data = [5, 12, 3, 20, 11, 9, 30];

// Part A
let final = [];
for (let i = 0; i < data.length; i++) {
  if (data[i] > 10) {
    final.push(data[i] * 2);
  }
}
console.log(`Loop:\nbefore: ${data}\nafter: ${final}\n`);

final = [];
final = (data.filter(n => n > 10)).map(n => n * 2);

console.log(`Pipe:\nbefore: ${data}\nafter: ${final}\n`);

// Part B
const numbers = [1, 2, 3, 4, 5, 6];
final = ((numbers.filter(n => n % 2 === 0)).map(n => n * 2)).reduce((sum, n) => sum + n, 0);

console.log(final + "\n");

// Part C
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];

console.log(users.some(n => n.age < 18));
console.log(users.find(n => n.id === 3));
users.forEach(n => {
  console.log(n.name);
});
console.log("\n");

// Part D
function repeat(n, action) {
  // TODO: call action(i) for i = 0..n-1
  for (let i = 0; i < n; i++) {
    action(i);
  }
}

repeat(3, console.log); // 0, 1, 2
repeat(4, i => console.log("Step " + i));
```
